```python
from nltk.tokenize import TreebankWordTokenizer

sentence = """The faster Harry got to the store, 
the faster Harry,the faster, would get home."""

tokenizer = TreebankWordTokenizer()
# Treebank tokenizer assumes that your document has already been segmented into 
# separate sentences

tokens = tokenizer.tokenize(sentence.lower())

tokens

from collections import Counter

bag_of_words = Counter(tokens)
bag_of_words
```

```python
# Naive Bayes
# need to pip install nlpia
from nlpia.data.loaders import get_data
movies = get_data('hutto_movies')

movies.head().round(2)

movies.describe().round(2)

import pandas as pd
pd.set_option('display.width', 75)

from nltk.tokenize import casual_tokenize

bow = []
from collections import Counter
for text in movies.text:
    bow.append(Counter(casual_tokenize(text)))

df_bows = pd.DataFrame.from_records(bow)
df_bows

df_bows = df_bows.fillna(0).astype(int)
df_bows

df_bows.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

nb = nb.fit(df_bows, movies.sentiment > 0)
```
